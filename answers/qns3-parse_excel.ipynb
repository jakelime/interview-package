```python
import pandas as pd
import re

def filter_alphanumeric(txt) -> str:
    pattern = re.compile(r"[A-Za-z0-9_]+")
    matches = pattern.findall(txt.replace(" ", "_"))
    return "".join(matches)

def get_headers(fpath) -> list:
    h1 = pd.read_excel(fpath, header=1, nrows=2, sheet_name=1)
    h2 = pd.read_excel(fpath, header=2, nrows=2, sheet_name=1)
    cols = []
    for a,b in zip(h1.columns, h2.columns):
        if "Unnamed" in b:
            cols.append(a)
        elif "Unnamed" in a:
            cols.append(b)
        else:
            if "US$" in b:
                cols.append(a)
            elif b=="REMARK":
                cols.append(b)
            else:
                cols.append(f"ERROR:{a},{b}")
    cols = [filter_alphanumeric(x) for x in cols]
    cols_remove_dup = []
    for i in range(len(cols)):
        if not cols_remove_dup:
            cols_remove_dup.append(cols[i])
            continue
        elif cols[i] in cols_remove_dup:
            cols_remove_dup.append(f"{cols[i]}_{cols_remove_dup.count(cols[i])}")
        else:
            cols_remove_dup.append(cols[i])
    return cols_remove_dup

def get_data(fpath) -> pd.DataFrame:
    start_row = 35
    df = pd.read_excel(
            fpath,
            sheet_name=1,
            header=start_row,
            nrows=start_row+10
        )
    return df

def parse_data(fpath) -> pd.DataFrame:
    h = get_headers(fpath)
    df = get_data(fpath)
    df.columns = h
    return df

```
